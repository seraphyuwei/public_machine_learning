```python
def split(split):
    for x in range(len(dataset)-1):
        if random.random() < split:
            trainingSet.append(dataset.loc[x])
        else:
            testSet.append(dataset.loc[x])
```

```python
# define the euclideanDistance function as follows
def euclidist(p1,p2):
    dist = 0
    for l in range(len(p1)-1):
        dist += pow(p1[l] - p2[l],2)
    dist = math.sqrt(dist)
    return dist
```

```python
def getneighbours(population,sample,k):
    distance = []
    for i in range(len(population)):
        dist = euclidist(population[i],sample)
        distance.append((population[i],dist))
    distance.sort(key=operator.itemgetter(1))
    
    neighbours = []
    for n in range(k):
        neighbours.append(distance[n][0])
    
    return neighbours
```

```python
def predict(neighbours):
    votes = {}
    for r in range(len(neighbours)):
        response = neighbours[r][-1]
        if response in votes:
            votes[response] += 1
        else:
            votes[response] = 1
    sorted_votes = sorted(votes.items(),key=operator.itemgetter(1),reverse=True)
    
    return sorted_votes[0][0]
```

```python
def getAccuracy(test,prediction):
    correct = 0
    for i in range(len(test)):
        if test[i][-1] is prediction[i]:
            correct += 1
    return correct/len(test) * 100
```
